```python
import numpy as np

file_path = 'source.txt'  # 7885693428401 < PART ONE
                          # 272500680812586 < PART TWO too low
                          # 348507745257079 < PART TWO too high
                          # 348360680516005 < CORRECT   (dif = 147 064 741 074)

with open(file_path, 'r') as file:
    lines = file.readlines()

full = []
for k,l in enumerate(lines):
    full.append(l.strip("\n"))

full_text= "".join(full)

sample =   ["190: 10 19",
"3267: 81 40 27",
"83: 17 5",
"156: 15 6",
"7290: 6 8 6 15",
"161011: 16 10 13",
"192: 17 8 14",
"21037: 9 7 18 13",
"292: 11 6 16 20"]

sample_text = "".join(sample)

# UTILISATION DU FICHIER AU LIEU DU SAMPLE
#sample=full
#sample_text=full_text
```

ON CREE NOTRE LISTE D'ELEMENTS

```python
lignes = []
for s in sample:
    total = s.split(":")[0]
    #numbers= [int(x) for x in s.split(":")[1].strip(" ").split(" ")]
    numbers = s.split(":")[1].strip(" ").split(" ")
    lignes.append([total,numbers])
    
```

```python
def test(k,v):
    if v[0]>k:
        return False
    if len(v)==1:
        return v[0]==k
    if(test(k, [v[0]+v[1]]+v[2:])):
        return True
    if(test(k, [v[0]*v[1]]+v[2:])):
        return True
    if(test(k, [int(str(v[0])+str(v[1]))]+v[2:])):
        return True
    return False
        

sum=0

for l in lignes:
    if(test(int(l[0]),l[1])):
        sum+=int(l[0])
print(sum)
```

ON CREE TOUTES LES PERMUTATIONS (39 SEC / 4 MIN 30 POUR LE FICHIER FINAL [PART 1 ET 2])

```python
import itertools
from copy import deepcopy

#PARTIE 1
operators = ["+","*"]
#PARTIE 2
operators = ["+","*", "||"]

liste_toutes_les_permutations = []

max_trous = 0

for l in lignes:
    if len(l[1])>max_trous:
        max_trous=len(l[1])
print(f"Nombre de valeurs max: {max_trous}")
  
for i in range(1,max_trous):
    print(f"permutations en cours pour {i} opérateur(s)")
    permut = set()
    combs = list(itertools.combinations_with_replacement(operators, i))
    for comb in combs:
        permut.update(itertools.permutations(comb))
    list_permut = []
    for s in permut:
        list_permut.append(list(s))
    liste_toutes_les_permutations.append(list_permut)
```
