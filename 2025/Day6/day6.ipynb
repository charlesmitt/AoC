```python
import numpy as np

file_path = 'source.txt'  

with open(file_path, 'r') as file:
    lines = file.readlines()

full = []
for k,l in enumerate(lines):
    full.append(l)

full_text= "".join(full)

sample =   ["....#.....",  # RESULTAT = 41 STEPS
            ".........#",
            "..........",
            "..#.......",
            ".......#..",
            "..........",
            ".#..^.....",
            "........#.",
            "#.........",
            "......#..."]

sample_text = "".join(sample)

# UTILISATION DU FICHIER AU LIEU DU SAMPLE
sample=full
sample_text=full_text

array = np.array(list(sample_text)).reshape(len(sample),len(sample[0]))
```

```python
orientation = None # U/D/L/R 

#       v  h
# array[0][4]

i_max_h = len(sample[0])-1
i_max_v = len(sample)-1
```

Trouver le garde et la direction

```python
pos = []
p = np.where(array ==">")
if len(p[0])>0:
    pos = p
    orientation="R"

p = np.where(array =="<")
if len(p[0])>0:
    pos = p
    orientation="L"

p = np.where(array =="^")
if len(p[0])>0:
    pos = p
    orientation="U"

p = np.where(array =="v")
if len(p[0])>0:
    pos = p
    orientation="D"
```

Les emplacements des obstacles

```python
obstacles = np.where(array =="#")
list_obstacles = []
for i,v in enumerate(obstacles[0]):
    list_obstacles.append([v,obstacles[1][i]])
```

PARTIE 1

```python
steps=[]
x_actuelle = pos

x= x_actuelle[0][0]
y= x_actuelle[1][0]

while((x<i_max_h) and (y<i_max_v) and (x>=0) and (y>=0)):
    match orientation:
        case "U":
            x_next = [x-1,y]
        case "D":
            x_next = [x+1,y]
        case "L":
            x_next = [x,y-1]
        case "R":
            x_next = [x,y+1]

    if x_next not in list_obstacles:
        if x_next not in steps:
            steps.append(x_next)
        x_actuelle=x_next
    else:
        match orientation:
            case "U":
                orientation="R"
            case "D":
                orientation="L"
            case "L":
                orientation="U"
            case "R":
                orientation="D"
    x= x_actuelle[0]
    y= x_actuelle[1]

print(len(steps))
```

PARTIE 2

```python
count = 0
liste_nouveaux_obstacles = []

# JUSTE POUR SAVOIR OU ON EN EST CAR L'ALGO PREND 15 MINUTES LOL
number = 1
nb_steps=len(steps)

# ON PARCOURT TOUTES LES CASES ET ON RAJOUTE UN OBSTACLE A CHAQUE FOIS
for s in steps:
    print(f"Obstacle {number}/{nb_steps}")
    iv = s[0]
    ih = s[1]
    
    # ON REINITIALISE LE LABYRINTHE
    array = np.array(list(sample_text)).reshape(len(sample),len(sample[0]))

    # ON RELOCALISE LE GARDE ET SA DIRECTION
    pos = []
    p = np.where(array ==">")
    if len(p[0])>0:
        pos = p
        orientation="R"

    p = np.where(array =="<")
    if len(p[0])>0:
        pos = p
        orientation="L"

    p = np.where(array =="^")
    if len(p[0])>0:
        pos = p
        orientation="U"

    p = np.where(array =="v")
    if len(p[0])>0:
        pos = p
        orientation="D"

    # ON RAJOUTE UN OBSTACLE ET ON REFAIT LA LISTE DES OBSTACLES 
    if (array[iv][ih]!= "#") and (array[iv][ih]!= "^"):
        array[iv][ih]="#"

        #print(f"Nouvel obstacle ({iv},{ih})")
        obstacles = np.where(array =="#")
        list_obstacles = []
        for i,v in enumerate(obstacles[0]):
            list_obstacles.append([v,obstacles[1][i]])

    #########################################

        steps=[]
        step_count=0
        x_actuelle = pos            

        x= x_actuelle[0][0]
        y= x_actuelle[1][0]
        loop = False
        while((x<i_max_h) and (y<i_max_v) and (x>=0) and (y>=0) and loop==False):
            
            # ON DEFINIT LA PROCHAINE CASE EN FONCTION DE L'ORIENTATION DU GARDE
            match orientation:
                case "U":
                    x_next = [x-1,y]
                case "D":
                    x_next = [x+1,y]
                case "L":
                    x_next = [x,y-1]
                case "R":
                    x_next = [x,y+1]

            # SI LA PROCHAINE CASE N'EST PAS UN OBSTACLE, ON DEPLACE LE GARDE
            if x_next not in list_obstacles:        
                x_actuelle=x_next
                step_count +=1
                x_next.append(orientation)
                
                if x_next in steps:
                    liste_nouveaux_obstacles.append([iv,ih])
                    count+=1
                    loop = True
                else:
                    steps.append(x_next)

            
            # SINON, ON TOURNE LE GARDE
            else:
                match orientation:
                    case "U":
                        orientation="R"
                    case "D":
                        orientation="L"
                    case "L":
                        orientation="U"
                    case "R":
                        orientation="D"

            # ON MET A JOUR LA POSITION 
            x= x_actuelle[0]
            y= x_actuelle[1]
        print(step_count)
        number += 1
```
